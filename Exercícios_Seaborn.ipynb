```
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```
sns.set_style('whitegrid')
sns.set_palette(sns.color_palette("Accent"))
```

```
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
titanic = pd.read_csv(url)
titanic.head()
```

```
del titanic['PassengerId']
```

```
titanic.head()
```

```
sns.jointplot(x = "Fare", y = "Age", data=titanic)
```

```
sns.distplot(titanic['Fare'], kde=False, color='red')
plt.xlim(0,500)
```

```
sns.boxplot(x='Pclass', y='Age', data=titanic, palette=sns.color_palette("pastel"))
```

```
sns.swarmplot(x='Pclass', y='Age', data=titanic)
```

```
sns.countplot(titanic['Sex'])
```

```
sns.heatmap(titanic.corr(), cmap="coolwarm")
```

```
fg = sns.FacetGrid(data=titanic, col="Sex")
fg = fg.map(plt.hist,"Age")
plt.xlim(left=0)
```
