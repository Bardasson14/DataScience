```
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```
import matplotlib.pyplot as plt
%matplotlib inline
```

```
#Exercício 1
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_xlim([0,100])
ax.set_ylim([0,200])
```

```
#Exercício 2
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2, 0.5, 0.2,0.2])
ax1.plot(x,y,'r')
ax2.plot(x,y,'r')
```

```
#Exercício 3
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([.2,.5,.4,.4])
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)
ax1.set_xlabel('x')
ax2.set_title('zoom')
ax1.plot(x,z)
ax2.plot(x,y)
```

```
#Exercício 4
fig, ax = plt.subplots(nrows=1, ncols=2)
plot1 = plt.subplot(1,2,1)
plot1.set_xlim([0,100])
plot1.set_ylim([0,200])
plot1.plot(x,y,color = "b", ls="--", lw = 3)

plot2 = plt.subplot(1,2,2)
plot2.set_xlim([0,100])
plot2.set_ylim([0,10000])
plot2.plot(x, z, color="r", lw=3)

fig.tight_layout()
```
